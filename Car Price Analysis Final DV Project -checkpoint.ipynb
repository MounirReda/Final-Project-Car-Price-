>**Detect and remove outliers**

```python
car_df_numeric = car_df.select_dtypes(exclude=object)
```

***In this step,we can detect outliers from the dataset because its cause nosie in data.
Steps to perform Outlier Detection by identifying the lowerbound and upperbound of the data:***
1. Arrange your data in ascending order
2. Calculate Q1 ( the first Quarter)
3. Calculate Q3 ( the third Quartile)
4. Find IQR = (Q3 - Q1)
5. Find the lower Range = Q1 -(1.5 * IQR)
6. Find the upper Range = Q3 + (1.5 * IQR)
***Once you get the upperbound and lowerbound, all you have to do is to delete any values which is less than lowerbound or greater than upperbound.***

```python
car_df_numeric=car_df.select_dtypes(exclude='object')
for col in car_df_numeric:
    q1=car_df[col].quantile(0.25)
    q3=car_df[col].quantile(0.75)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    outlier=((car_df_numeric[col]>high)|(car_df_numeric[col]<low)).sum()
    total=car_df_numeric[col].shape[0]
    print(f"Total Outliers in {col} are :{outlier}-{round(100*(outlier)/total,2)}%")
    if outlier >0:
        car_df=car_df.loc[(car_df[col]<=high) & (car_df[col]>=low)]
```

```python
#Plotting Graphs after treating outliers of continous features
for col in car_df_numeric:
    fig, ax =plt.subplots(1,2, constrained_layout=True)
    fig.set_size_inches(20, 6)
    sns.distplot(car_df[col], ax=ax[0],color='red',kde=True).set(title="Distplot")
    sns.boxplot(car_df[col], ax=ax[1],color='red').set(title="Boxplot")
    plt.suptitle(f'{col.title()} (after handling outliers)',weight='bold')
    fig.show()
```

```python
x = car_df.drop('Price',axis=1)   
y = car_df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size=0.2,random_state=42)
```

```python
Algorithm= ['LinearRegression','DecisionTreeRegressor','RandomForestRegressor','GradientBoostingRegressor','XGBRegressor','SVR']
R2=[]
RMSE=[]
```

```python


def train_model(model):
    model.fit(x_train, y_train)
    pre = model.predict(x_test)
    r2 = r2_score(y_test, pre) # compare between masked and the value that the model predicted 
    rmse = np.sqrt(mean_squared_error(y_test, pre))
    R2.append(r2)
    RMSE.append(rmse)
    score = model.score(x_test, y_test)
    return {'Model': model.__class__.__name__, 'R2 Score': r2, 'RMSE': rmse, 'Score': score, 'Predictions': pre}

    
    
```

```python
model1 = LinearRegression()
model2 = DecisionTreeRegressor ()
model3 = RandomForestRegressor ()
model4 = GradientBoostingRegressor ()
model5 = XGBRegressor ()
model6 = SVR ()
```

```python
train_model(model1)
train_model(model2)
train_model(model3)
train_model(model4)
train_model(model5)
train_model(model6)
```
